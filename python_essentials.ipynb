```
# Using a dictionary to store values in a (key,value):
# The key should be an immutable dtaa type such as a string or a tuple:

# Normal method:
teams = ['Manchester', 'Liverpool', 'Arsenal', 'Chelsea', 'Mancity', 'Tottenham', 'Barcelona','Dortmund']

clubs = {}

for team in teams:
  clubInitials = team[0] + team[-2:]
  
  # If the key exist with the club initials then just append the team: 
  if clubInitials in clubs.keys():
    clubs[clubInitials].append(team)
  else:
    # Create a new list and key referencing in clubs:
    clubs[clubInitials] = [team]
    

print(clubs)
```
