```
# Pandas:
import numpy as np
import pandas as pd
from pandas import Series, DataFrame


# Series: 1D array tjta behaves like a dictionary:
obj = pd.Series([4 ,7, -5, 3])
print(obj)


# Keys are by default numbers:
print('Keys: ' + str(obj.index))
print('Values:' + str(obj.values))

# Assign char keys to values via Series:
obj2 = pd.Series([4,7,-5,3], index=['d','b','a','c'])
print(obj2)

print(obj2.index)

# Select values using normal array selection:
print('a: ' + str(obj2['a']))
print('d: '+ str(obj2['d']))

# Can swap filed order temporarily:
print(obj2[['c','a','d']])
print(obj2)


# Can filter to generate a boolean array:
print(obj2 > 0)

# Can update values via scalar mult:
print(obj2 * 2)

# Can perform math funcions:
np.exp(obj2)

# Can be used in many of the same circumstances in which we use a dict:
print('b' in obj2)

# We can create  Series from a dict:
sdata = {'Ohio':35000, 'Texas': 71000, 'Oregon': 16000, 'Utah': 5000}
obj3 = pd.Series(sdata)
print(obj3)


# We can override the order of of the dict keys by manually ordering:
# CA: NAN - Does not exist in sdata
states = ['California', 'Ohio', 'Oregon', 'Texas']
obj4 = pd.Series(sdata, index=states)
print(obj4)

# isNull() / notNull functions provide Boolean output:
print(pd.isnull(obj4))

# Also a instance method on the object itself:
print(obj4.isnull())

# Can combine series via arithmetic operations
# which operates according to the key/index value
# Works similar to a JOIN in SQL by field
print()
print('Obj3: ')
print(obj3)
print('Obj4:' )
print(obj4)
print('Sum of obj3 + obj4: ')
print(obj3 + obj4)


# SERIES objects have both name and index attributes:
obj4.name = 'population'

obj4.index.name = 'state'

print(obj4)

# A series indexes can be altered inplace:
print('Default numerical indexes can be replaced by string indexes INPLACE by assignment:')
print(obj)

obj.index = ['Bob','Steve','Jeff', 'Ryan']
print('Updated indexes: ')
print(obj)
```
